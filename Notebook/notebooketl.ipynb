```python
df = pd.read_csv("ecommerce_product_dataset.csv")
```

```python
df.head()
```

```python
#checking for missing values
df = df.dropna()

# Convert 'DateAdded' to datetime
df['DateAdded'] = pd.to_datetime(df['DateAdded'])
# Save cleaned DataFrame to CSV
output_path = r"C:\Users\Administrator\cleaned_products.csv"
df.to_csv(output_path, index=False)

print(f"Cleaned data saved to: {output_path}")



# Check cleaned data
df.describe()
```

```python
import sqlite3

# Connect/create database
conn = sqlite3.connect('ecommerce_data.db')

# Save DataFrame to SQLite table
df.to_sql('products', conn, if_exists='replace', index=False)
```

```python
df_items = pd.read_sql_query("SELECT * FROM products;", conn)
df_items.head
```

```python
# Close connection
conn.close()
```
